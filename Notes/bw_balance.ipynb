```
def is_image_good(img):
  directory = img
  img = cv2.imread(img,0)
  hist,bins = np.histogram(img.ravel(),256,[0,256])
  total = np.sum(hist)
  
  black = 0
  white = 0

  for i in range(0, 128):
    black += hist[i]
  
  for i in range(128, 256):
    white += hist[i]
  
  if black/total > 0.75:
    status = False
  elif white/total > 0.75:
    status = False
  else:
    status = True
    
  print(total, white, black, white/total, black/total, status)
  #return status
```
